```python
%maxtplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
fig = plt.figure
ax = plt.axes()

x = np.linspace(0,10,100)
ax.plot(x,np.sin(x),'--c',label='sin(x)') #青色虚线
ax.plot(x,np.cos(x),'-.k',label='cos(x)')  #黑色点划线

#设置坐标轴上下限
plt.xlim(-1,11)
plt.ylim(-2,2)
#或者使用 plt.axis([-1,11,2,2])

plt.title('Some Cruves')
plt.xlabel('x')

plt.legend()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
plt.scatter(x,y,marker='o')
```

```python
#scatter可以设置每个点的属性
rng = np.random.RandomState(42)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000*rng.rand(100)

plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='viridis')
plt.colorbar() #显示颜色条
```
