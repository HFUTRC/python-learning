```python
df = pd.DataFrame(data=np.random.randint(0,150,size=(100,5)),index=np.arange(100,200),columns=['Python','Math','Eng','Physic','Chem'])
df
```

```python
df.isnull().any()
```

```python
for i in range(50):
    # 行索引
    index = np.random.randint(100,200,size =1)[0]

    cols = df.columns

    # 列索引
    col = np.random.choice(cols)

    df.loc[index,col] = None
```

```python
for i in range(20):
    # 行索引
    index = np.random.randint(100,200,size =1)[0]

    cols = df.columns

    # 列索引
    col = np.random.choice(cols)

#     not a number 不是一个数
    df.loc[index,col] = np.NAN
```
